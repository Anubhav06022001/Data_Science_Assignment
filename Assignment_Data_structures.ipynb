2)Stack would be a more practical choice than a list for data storage and retrieval is in the management of the call stack in a programming language interpreter or compiler.
Stack is more practical because of:
->LIFO
->Efficient Push and Pop Operations
->Limited Access
->Automatic Memory Management
->Recursion Support



```python
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    # Recursive case
    else:
        return n * factorial(n - 1)


print("Factorial of 5:", factorial(5))

#Each recursive call is pushed onto the call stack, and when a base case is reached, the corresponding stack frame is popped off the stack. This process continues until the initial call to factorial(5) returns the final result, which is printed as the factorial of 5.
```

3)The primary advantage of using sets in Python is their ability to efficiently store unique elements and perform set operations such as intersection, union, difference, and symmetric difference. Sets are implemented using hash tables, which provide near-constant time complexity for common operations like insertion, deletion, and membership testing.

some problem solving scenarios are:

a)Removing Duplicates
b)Membership Testing
c)Intersection, Union, and Difference Operations
d)Counting Unique Elements
e)Filtering Data
f)Graph Algorithms


```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Union of sets
union_set = set1.union(set2)
print("Union:", union_set)

# Intersection of sets
intersection_set = set1.intersection(set2)
print("Intersection:", intersection_set)

# Difference of sets
difference_set = set1.difference(set2)
print("Difference:", difference_set)

# Symmetric difference of sets
symmetric_difference_set = set1.symmetric_difference(set2)
print("Symmetric Difference:", symmetric_difference_set)
```

5)In lists each item is accessed by index value while in dictionary each item is stored in the form of key value pair.

lists allow duplication but not dictionary

Lists are suitable for scenarios where maintaining the order of elements and accessing them by their position is important. They are commonly used for storing and manipulating sequences of data, such as lists of numbers, strings, or objects.

Dictionaries are ideal for scenarios where you need to associate unique keys with corresponding values and quickly look up values based on their keys. They are commonly used for mappings, such as storing user information, configuration settings, or any data that requires efficient key-based access.

```python
#list example
my_lis=[1,2,3,4,5]

#dictionary example
my_dict={"name":"Anubhav","course":"Data Science","Age":23}
```

```python
print(my_lis[1])
```

```python
print(my_dict["name"])
```

```python
#list allows duplication
my_lis.append(5)
print(my_lis)
```

```python
#dictionary does not allow duplication
my_dict.update({"name":"Anubhav"})
print(my_dict)
```
