1) len() function is used to determine the length of a given sequence or collection      
            
 -->It can be used with various data types such as strings, lists, tuples, dictionaries, etc.            

-->When used with strings specifically, len() returns the number of characters in the string, including spaces, punctuation, and special characters.

```python
a="Anubhav"
b=len(a)
print("The length of string is",b)
```

2) In Python, we can concatenate two or more strings using the + operator or the str.join() method.


```python
a="Anubhav"
b=" Tripathi"
c=a+b
print(c)
```

```python
d=[a, b]
e="".join(d)
print(e)
```

3)string slicing is a technique used to extract a portion of a string by specifying a range of indices.

This allows you to access specific parts of a string without modifying the original string.

```python
a="Anubhav Tripathi"
slicing=a[0:7]
print(slicing)
```

4) In Python, we can change the case of a string using the upper() and lower() methods. These methods return a new string with the case changed according to the method's name.

```python
#to change to lower case
b=a.lower()
print(b)

#to change to upper case
c=a.upper()
print(c)
```

5) split() method is used to split a string into a list of substrings based on a specified delimiter.

```python
f=a.split()
print(f)
```
